```python
import pandas as pd
import numpy as np

np.random.seed(42)

# ==========================================
# LOAD DIMENSIONS
# ==========================================

customers = pd.read_csv("dim_customer.csv")
branches = pd.read_csv("dim_branch.csv")
dates = pd.read_csv("dim_date.csv")
products = pd.read_csv("dim_product.csv")

# ==========================================
# DEPOSIT PRODUCTS ONLY
# ==========================================

deposit_products = products[
    products["Product_Category"] == "Deposit"
].copy()

# ==========================================
# LOOKUPS
# ==========================================

income_lookup = customers.set_index(
    "Customer_ID"
)["Income_Band"].to_dict()

segment_lookup = customers.set_index(
    "Customer_ID"
)["Segment_ID"].to_dict()

product_lookup = products.set_index(
    "Product_ID"
)["Product_Name"].to_dict()

interest_lookup = products.set_index(
    "Product_ID"
)["Interest_Rate"].to_dict()

# ==========================================
# CONFIG
# ==========================================

N = 500000

# ==========================================
# RANDOM KEYS
# ==========================================

customer_ids = np.random.choice(
    customers["Customer_ID"],
    N
)

branch_ids = np.random.choice(
    branches["Branch_ID"],
    N
)

date_keys = np.random.choice(
    dates["Date_Key"],
    N
)

# 8 Deposit Products
product_ids = np.random.choice(

    deposit_products["Product_ID"],

    N,

    p=[
        0.30,  # Savings Account
        0.12,  # Current Account
        0.15,  # Fixed Deposit
        0.08,  # Islamic Deposit
        0.10,  # SME Current Account
        0.08,  # SME Term Deposit
        0.07,  # Corporate Current Account
        0.10   # Priority Savings
    ]

)

# ==========================================
# BALANCE GENERATION
# ==========================================

balances = []

for cid, pid in zip(
    customer_ids,
    product_ids
):

    band = income_lookup.get(
        cid,
        "Middle"
    )

    segment = segment_lookup.get(
        cid,
        1
    )

    product = product_lookup.get(
        pid,
        ""
    )

    # Base balance

    if band == "Low":

        balance = np.random.lognormal(
            9,
            0.8
        )

    elif band == "Middle":

        balance = np.random.lognormal(
            10,
            0.8
        )

    elif band == "Upper Middle":

        balance = np.random.lognormal(
            11,
            0.7
        )

    else:

        balance = np.random.lognormal(
            12,
            0.6
        )

    # Segment multiplier

    if segment == 2:

        balance *= np.random.uniform(
            3,
            8
        )

    elif segment == 3:

        balance *= np.random.uniform(
            20,
            60
        )

    elif segment == 4:

        balance *= np.random.uniform(
            5,
            15
        )

    # Product multiplier

    if "Fixed Deposit" in product:

        balance *= np.random.uniform(
            2,
            5
        )

    elif "Term Deposit" in product:

        balance *= np.random.uniform(
            3,
            6
        )

    elif "Corporate Current" in product:

        balance *= np.random.uniform(
            5,
            15
        )

    balances.append(balance)

balances = np.array(
    balances
)

# ==========================================
# TRANSACTION FLOWS
# ==========================================

deposit_amount = (

    balances *

    np.random.uniform(
        0.01,
        0.15,
        N
    )

)

withdrawal_amount = (

    balances *

    np.random.uniform(
        0.005,
        0.12,
        N
    )

)

# ==========================================
# INTEREST CALCULATION
# ==========================================

interest_paid = []

for bal, pid in zip(
    balances,
    product_ids
):

    annual_rate = interest_lookup.get(
        pid,
        5
    )

    monthly_interest = (

        bal *

        (annual_rate / 100)

        / 12

    )

    interest_paid.append(
        monthly_interest
    )

interest_paid = np.array(
    interest_paid
)

# ==========================================
# ACCOUNT STATUS
# ==========================================

status = np.random.choice(

    [
        "Active",
        "Dormant",
        "Closed"
    ],

    N,

    p=[
        0.90,
        0.08,
        0.02
    ]

)

# ==========================================
# ACCOUNT NUMBER
# ==========================================

account_no = []

counters = {

    "RET":10000001,
    "SME":20000001,
    "COR":30000001,
    "PRI":40000001

}

for pid in product_ids:

    product = product_lookup.get(
        pid,
        ""
    )

    if "SME" in product:

        prefix = "SME"

    elif "Corporate" in product:

        prefix = "COR"

    elif "Priority" in product:

        prefix = "PRI"

    else:

        prefix = "RET"

    account_no.append(

        f"{prefix}{counters[prefix]}"

    )

    counters[prefix] += 1

# ==========================================
# OPENING DATE
# ==========================================

opening_dates = pd.to_datetime(

    np.random.choice(

        pd.date_range(
            "2020-01-01",
            "2025-12-31"
        ),

        N

    )

)

# ==========================================
# FACT TABLE
# ==========================================

fact_deposit = pd.DataFrame({

    "Deposit_ID":
        range(
            1,
            N + 1
        ),

    "Date_Key":
        date_keys,

    "Customer_ID":
        customer_ids,

    "Branch_ID":
        branch_ids,

    "Product_ID":
        product_ids,

    "Account_No":
        account_no,

    "Opening_Date":
        opening_dates,

    "Deposit_Balance":
        balances.round(2),

    "Deposit_Amount":
        deposit_amount.round(2),

    "Withdrawal_Amount":
        withdrawal_amount.round(2),

    "Interest_Paid":
        interest_paid.round(2),

    "Account_Status":
        status,

    "Snapshot_Date":
        pd.Timestamp.today().date()

})

# ==========================================
# EXPORT
# ==========================================

fact_deposit.to_csv(
    "fact_deposit.csv",
    index=False
)

print(
    fact_deposit.head()
)

print(
    fact_deposit.shape
)
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# ==========================================================
# LOAD DIMENSIONS
# ==========================================================

customers = pd.read_csv("dim_customer.csv")
branches = pd.read_csv("dim_branch.csv")
dates = pd.read_csv("dim_date.csv")
products = pd.read_csv("dim_product.csv")
segments = pd.read_csv("dim_segment.csv")

# ==========================================================
# CONFIG
# ==========================================================

N = 150000

# ==========================================================
# LOAN PRODUCTS ONLY
# ==========================================================

loan_products = products[
    products["Product_Category"] == "Loan"
].copy()

# ==========================================================
# LOOKUP TABLES
# ==========================================================

income_lookup = customers.set_index(
    "Customer_ID"
)["Income_Band"].to_dict()

segment_lookup = customers.set_index(
    "Customer_ID"
)["Segment_ID"].to_dict()

product_lookup = products.set_index(
    "Product_ID"
)["Product_Name"].to_dict()

interest_lookup = products.set_index(
    "Product_ID"
)["Interest_Rate"].to_dict()

# ==========================================================
# RANDOM DIMENSION KEYS
# ==========================================================

customer_ids = np.random.choice(
    customers["Customer_ID"],
    N
)

branch_ids = np.random.choice(
    branches["Branch_ID"],
    N
)

date_keys = np.random.choice(
    dates["Date_Key"],
    N
)

# ==========================================================
# LOAN PRODUCT DISTRIBUTION
# ==========================================================

loan_product_weights = {

    "Home Loan":0.22,

    "Personal Loan":0.20,

    "Auto Loan":0.08,

    "SME Loan":0.18,

    "Business Overdraft":0.10,

    "Working Capital Loan":0.12,

    "Corporate Term Loan":0.10

}

loan_products = loan_products[
    loan_products["Product_Name"].isin(
        loan_product_weights.keys()
    )
].copy()

loan_products["Weight"] = loan_products[
    "Product_Name"
].map(
    loan_product_weights
)

product_ids = np.random.choice(

    loan_products["Product_ID"],

    N,

    p=loan_products["Weight"]

)

# ==========================================================
# VERIFY PRODUCT DISTRIBUTION
# ==========================================================

print("\nLoan Product Distribution\n")

print(

    loan_products[
        [
            "Product_ID",
            "Product_Name",
            "Business_Unit",
            "Interest_Rate",
            "Weight"
        ]
    ]

)

print("\nRows To Generate :", N)
# ==========================================================
# ENTERPRISE LOAN GENERATION LOGIC
# ==========================================================

loan_amounts = []
interest_rates = []
loan_terms = []
outstanding_balances = []
emi_amounts = []
interest_income = []
days_past_due = []
loan_statuses = []
collateral_values = []
ltv_ratios = []

for cid, pid in zip(customer_ids, product_ids):

    pname = product_lookup.get(pid, "")
    band = income_lookup.get(cid, "Middle")
    segment = segment_lookup.get(cid, 1)

    # ----------------------------------------
    # HOME LOAN
    # ----------------------------------------

    if pname == "Home Loan":

        if band == "Low":

            amount = np.random.uniform(
                800000,
                3000000
            )

        elif band == "Middle":

            amount = np.random.uniform(
                2000000,
                7000000
            )

        elif band == "Upper Middle":

            amount = np.random.uniform(
                5000000,
                15000000
            )

        else:

            amount = np.random.uniform(
                10000000,
                30000000
            )

        term = np.random.choice(
            [120,180,240,300]
        )

        collateral = amount * np.random.uniform(
            1.10,
            1.60
        )

    # ----------------------------------------
    # SME LOAN
    # ----------------------------------------

    elif pname == "SME Loan":

        amount = np.random.uniform(
            500000,
            20000000
        )

        term = np.random.choice(
            [24,36,48,60]
        )

        collateral = amount * np.random.uniform(
            1.00,
            1.50
        )

    # ----------------------------------------
    # PERSONAL LOAN
    # ----------------------------------------

    elif pname == "Personal Loan":

        if band == "Low":

            amount = np.random.uniform(
                50000,
                200000
            )

        elif band == "Middle":

            amount = np.random.uniform(
                200000,
                600000
            )

        else:

            amount = np.random.uniform(
                600000,
                1000000
            )

        term = np.random.choice(
            [12,24,36,48,60]
        )

        collateral = 0

    # ----------------------------------------
    # AUTO LOAN
    # ----------------------------------------

    else:

        amount = np.random.uniform(
            500000,
            3000000
        )

        term = np.random.choice(
            [24,36,48,60]
        )

        collateral = amount * np.random.uniform(
            1.05,
            1.30
        )

    # ----------------------------------------
    # SEGMENT ADJUSTMENT
    # ----------------------------------------

    if segment == 2:

        amount *= np.random.uniform(
            1.20,
            1.80
        )

    elif segment == 4:

        amount *= np.random.uniform(
            1.10,
            1.50
        )

    # ----------------------------------------
    # INTEREST RATE
    # ----------------------------------------

    rate = interest_lookup.get(
        pid,
        10
    )

    # ----------------------------------------
    # OUTSTANDING BALANCE
    # ----------------------------------------

    outstanding = amount * np.random.uniform(
        0.35,
        1.00
    )

    # ----------------------------------------
    # EMI
    # ----------------------------------------

    monthly_rate = rate / 100 / 12

    emi = (

        amount *

        monthly_rate *

        (1 + monthly_rate) ** term

    ) / (

        ((1 + monthly_rate) ** term) - 1

    )

    # ----------------------------------------
    # MONTHLY INTEREST INCOME
    # ----------------------------------------

    interest_inc = outstanding * monthly_rate

    # ----------------------------------------
    # DAYS PAST DUE
    # ----------------------------------------

    dpd = np.random.choice(

        [0,30,60,90,120],

        p=[
            0.85,
            0.07,
            0.04,
            0.03,
            0.01
        ]

    )

    if dpd >= 90:

        status = "NPL"

    elif dpd > 0:

        status = "Delinquent"

    else:

        status = "Current"

    # ----------------------------------------
    # LTV
    # ----------------------------------------

    if collateral > 0:

        ltv = (
            amount /
            collateral
        ) * 100

    else:

        ltv = np.nan

    # ----------------------------------------
    # STORE
    # ----------------------------------------

    loan_amounts.append(amount)

    interest_rates.append(rate)

    loan_terms.append(term)

    outstanding_balances.append(outstanding)

    emi_amounts.append(emi)

    interest_income.append(
        interest_inc
    )

    days_past_due.append(
        dpd
    )

    loan_statuses.append(
        status
    )

    collateral_values.append(
        collateral
    )

    ltv_ratios.append(
        ltv
    )

print("Loan generation completed.")
# ==========================================================
# DISBURSEMENT DATE
# ==========================================================

disbursement_dates = pd.to_datetime(

    np.random.choice(

        dates["Full_Date"],

        N

    )

)

# ==========================================================
# MATURITY DATE
# ==========================================================

maturity_dates = []

for disb_date, term in zip(

    disbursement_dates,

    loan_terms

):

    maturity_dates.append(

        disb_date +

        pd.DateOffset(

            months=int(term)

        )

    )

# ==========================================================
# LOAN ACCOUNT NUMBER
# ==========================================================

loan_account_no = []

counters = {

    "RET": 50000001,
    "SME": 60000001,
    "PRI": 70000001

}

for cid in customer_ids:

    segment = segment_lookup.get(cid, 1)

    if segment == 2:

        prefix = "SME"

    elif segment == 4:

        prefix = "PRI"

    else:

        prefix = "RET"

    loan_account_no.append(

        f"{prefix}L{counters[prefix]}"

    )

    counters[prefix] += 1

# ==========================================================
# SNAPSHOT DATE
# ==========================================================

snapshot_date = pd.Timestamp.today().date()

# ==========================================================
# ROUND NUMERIC FIELDS
# ==========================================================

loan_amounts = np.round(
    loan_amounts,
    2
)

outstanding_balances = np.round(
    outstanding_balances,
    2
)

emi_amounts = np.round(
    emi_amounts,
    2
)

interest_rates = np.round(
    interest_rates,
    2
)

interest_income = np.round(
    interest_income,
    2
)

collateral_values = np.round(
    collateral_values,
    2
)

ltv_ratios = np.round(
    ltv_ratios,
    2
)

# ==========================================================
# FINAL VALIDATION
# ==========================================================

assert len(customer_ids) == N
assert len(branch_ids) == N
assert len(product_ids) == N
assert len(loan_amounts) == N
assert len(outstanding_balances) == N
assert len(emi_amounts) == N
assert len(interest_rates) == N
assert len(interest_income) == N
assert len(days_past_due) == N
assert len(loan_statuses) == N
assert len(collateral_values) == N
assert len(ltv_ratios) == N
assert len(disbursement_dates) == N
assert len(maturity_dates) == N
assert len(loan_account_no) == N

print("Part 3 Completed Successfully.")
# ==========================================================
# FACT LOAN TABLE
# ==========================================================

fact_loan = pd.DataFrame({

    "Loan_ID":
        range(
            1,
            N + 1
        ),

    "Date_Key":
        date_keys,

    "Customer_ID":
        customer_ids,

    "Branch_ID":
        branch_ids,

    "Product_ID":
        product_ids,

    "Loan_Account_No":
        loan_account_no,

    "Loan_Amount":
        loan_amounts,

    "Outstanding_Balance":
        outstanding_balances,

    "EMI_Amount":
        emi_amounts,

    "Interest_Rate":
        interest_rates,

    "Interest_Income":
        interest_income,

    "Loan_Term_Months":
        loan_terms,

    "Collateral_Value":
        collateral_values,

    "LTV_Ratio":
        ltv_ratios,

    "Disbursement_Date":
        disbursement_dates,

    "Maturity_Date":
        maturity_dates,

    "Days_Past_Due":
        days_past_due,

    "Loan_Status":
        loan_statuses,

    "Snapshot_Date":
        snapshot_date

})

# ==========================================================
# SORT TABLE
# ==========================================================

fact_loan = fact_loan.sort_values(

    by=[
        "Loan_ID"
    ]

).reset_index(

    drop=True

)

# ==========================================================
# DATA QUALITY CHECK
# ==========================================================

print("\n==============================")
print("FACT LOAN SUMMARY")
print("==============================")

print(f"Rows    : {len(fact_loan):,}")
print(f"Columns : {len(fact_loan.columns)}")

print("\nLoan Status Distribution")

print(

    fact_loan["Loan_Status"]

    .value_counts()

)

print("\nProduct Distribution")

print(

    fact_loan["Product_ID"]

    .value_counts()

)

print("\nMissing Values")

print(

    fact_loan.isnull()

    .sum()

)

print("\nFirst Five Rows")

print(

    fact_loan.head()

)

# ==========================================================
# EXPORT
# ==========================================================

fact_loan.to_csv(

    "fact_loan.csv",

    index=False

)

print("\n===================================")
print("fact_loan.csv Export Successful")
print("===================================")
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# ==========================================================
# LOAD FACT LOAN
# ==========================================================

fact_loan = pd.read_csv(
    "fact_loan.csv"
)

# ==========================================================
# SELECT NPL LOANS
# ==========================================================

npl_loans = fact_loan[

    fact_loan["Loan_Status"] == "NPL"

].copy()

# ==========================================================
# KEEP REQUIRED COLUMNS
# ==========================================================

npl_loans = npl_loans[

    [

        "Loan_ID",

        "Date_Key",

        "Customer_ID",

        "Branch_ID",

        "Product_ID",

        "Loan_Account_No",

        "Outstanding_Balance",

        "Interest_Income",

        "Days_Past_Due",

        "Loan_Status"

    ]

].copy()

# ==========================================================
# RESET INDEX
# ==========================================================

npl_loans.reset_index(

    drop=True,

    inplace=True

)

print(f"Total NPL Loans : {len(npl_loans):,}")

print("Part 1 Completed Successfully.")
# ==========================================================
# STAGE CLASSIFICATION
# ==========================================================

stages = []
provision_rates = []

for dpd in npl_loans["Days_Past_Due"]:

    if dpd < 90:

        stage = "Stage 2"
        rate = np.random.uniform(
            0.05,
            0.15
        )

    elif dpd < 180:

        stage = "Stage 3"
        rate = np.random.uniform(
            0.20,
            0.50
        )

    else:

        stage = "Stage 3 Severe"
        rate = np.random.uniform(
            0.60,
            1.00
        )

    stages.append(stage)
    provision_rates.append(rate)

# ==========================================================
# PROVISION
# ==========================================================

provisions = (

    npl_loans["Outstanding_Balance"]

    *

    provision_rates

)

# ==========================================================
# RECOVERY
# ==========================================================

recovery_rates = np.random.uniform(

    0.00,

    0.40,

    len(npl_loans)

)

recovery = (

    npl_loans["Outstanding_Balance"]

    *

    recovery_rates

)

# ==========================================================
# WRITE OFF
# ==========================================================

write_off = []

for stage, outstanding in zip(

    stages,

    npl_loans["Outstanding_Balance"]

):

    if stage == "Stage 3 Severe":

        if np.random.rand() < 0.20:

            write_off.append(

                outstanding *

                np.random.uniform(
                    0.30,
                    1.00
                )

            )

        else:

            write_off.append(0)

    else:

        write_off.append(0)

print("Part 2 Completed Successfully.")
# ==========================================================
# SNAPSHOT DATE
# ==========================================================

snapshot_date = pd.Timestamp.today().date()

# ==========================================================
# FACT NPL TABLE
# ==========================================================

fact_npl = pd.DataFrame({

    "NPL_ID":
        range(
            1,
            len(npl_loans)+1
        ),

    "Loan_ID":
        npl_loans["Loan_ID"],

    "Date_Key":
        npl_loans["Date_Key"],

    "Customer_ID":
        npl_loans["Customer_ID"],

    "Branch_ID":
        npl_loans["Branch_ID"],

    "Product_ID":
        npl_loans["Product_ID"],

    "Loan_Account_No":
        npl_loans["Loan_Account_No"],

    "Outstanding":
        npl_loans["Outstanding_Balance"].round(2),

    "Interest_Income":
        npl_loans["Interest_Income"].round(2),

    "Days_Past_Due":
        npl_loans["Days_Past_Due"],

    "NPL_Flag":
        1,

    "Stage":
        stages,

    "Provision_Rate":
        np.round(
            provision_rates,
            4
        ),

    "Provision":
        np.round(
            provisions,
            2
        ),

    "Recovery_Amount":
        np.round(
            recovery,
            2
        ),

    "Write_Off":
        np.round(
            write_off,
            2
        ),

    "Snapshot_Date":
        snapshot_date

})

fact_npl = fact_npl.sort_values(

    "NPL_ID"

).reset_index(

    drop=True

)

print("Part 3 Completed Successfully.")
# ==========================================================
# DATA QUALITY CHECK
# ==========================================================

print("\n===================================")
print("FACT NPL SUMMARY")
print("===================================")

print(f"Rows    : {len(fact_npl):,}")
print(f"Columns : {len(fact_npl.columns)}")

print("\nStage Distribution")

print(

    fact_npl["Stage"]

    .value_counts()

)

print("\nProvision Summary")

print(

    fact_npl["Provision"]

    .describe()

)

print("\nWrite Off Summary")

print(

    fact_npl["Write_Off"]

    .describe()

)

print("\nMissing Values")

print(

    fact_npl

    .isnull()

    .sum()

)

print("\nFirst Five Rows")

print(

    fact_npl.head()

)

# ==========================================================
# EXPORT
# ==========================================================

fact_npl.to_csv(

    "fact_npl.csv",

    index=False

)

print("\n===================================")
print("fact_npl.csv Export Successful")
print("===================================")

print(f"Final Shape : {fact_npl.shape}")
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# ==========================================================
# LOAD FACT TABLES
# ==========================================================

fact_loan = pd.read_csv(
    "fact_loan.csv"
)

fact_transaction = pd.read_csv(
    "fact_transaction.csv"
)

fact_npl = pd.read_csv(
    "fact_npl.csv"
)

# ==========================================================
# LOAN AGGREGATION
# ==========================================================

loan_agg = (

    fact_loan

    .groupby(

        [
            "Date_Key",
            "Branch_ID",
            "Product_ID"
        ]

    )

    .agg({

        "Outstanding_Balance":"sum",

        "Interest_Income":"sum"

    })

    .reset_index()

)

# ==========================================================
# TRANSACTION AGGREGATION
# ==========================================================

txn_agg = (

    fact_transaction

    .groupby(

        [
            "Date_Key",
            "Branch_ID",
            "Product_ID"
        ]

    )

    .agg({

        "Transaction_Fee":"sum"

    })

    .reset_index()

)

txn_agg.rename(

    columns={

        "Transaction_Fee":"Fee_Income"

    },

    inplace=True

)

# ==========================================================
# NPL AGGREGATION
# ==========================================================

npl_agg = (

    fact_npl

    .groupby(

        [
            "Date_Key",
            "Branch_ID"
        ]

    )

    .agg({

        "Provision":"sum"

    })

    .reset_index()

)

print("Part 1 Completed Successfully.")
# ==========================================================
# MERGE ALL SOURCES
# ==========================================================

profit = loan_agg.merge(

    npl_agg,

    on=[
        "Date_Key",
        "Branch_ID"
    ],

    how="left"

)

profit = profit.merge(

    txn_agg,

    on=[
        "Date_Key",
        "Branch_ID",
        "Product_ID"
    ],

    how="left"

)

# ==========================================================
# HANDLE MISSING VALUES
# ==========================================================

profit["Provision"] = (

    profit["Provision"]

    .fillna(0)

)

profit["Fee_Income"] = (

    profit["Fee_Income"]

    .fillna(0)

)

# ==========================================================
# PROVISION EXPENSE
# ==========================================================

profit["Provision_Expense"] = (

    profit["Provision"]

    *

    np.random.uniform(
        0.01,
        0.05,
        len(profit)
    )

)


# ==========================================================
# COMMISSION INCOME
# ==========================================================

profit["Commission_Income"] = (

    profit["Fee_Income"]

    *

    np.random.uniform(
        0.20,
        0.60,
        len(profit)

    )

)


# ==========================================================
# OPERATING EXPENSE
# ==========================================================

profit["Operating_Expense"] = (

    profit["Outstanding_Balance"]

    *

    np.random.uniform(
        0.0005,
        0.0015,
        len(profit)

    )

)


# ==========================================================
# GROSS PROFIT
# ==========================================================

profit["Gross_Profit"] = (

    profit["Interest_Income"]

    +

    profit["Fee_Income"]

    +

    profit["Commission_Income"]

    -

    profit["Operating_Expense"]

)



# ==========================================================
# TAX
# ==========================================================

profit["Tax"] = (

    (

        profit["Gross_Profit"]

        -

        profit["Provision_Expense"]

    )

    .clip(lower=0)

    *

    0.275

)



# ==========================================================
# NET PROFIT
# ==========================================================

profit["Net_Profit"] = (

    profit["Gross_Profit"]

    -

    profit["Provision_Expense"]

    -

    profit["Tax"]

)

# ==========================================================
# ROA
# ==========================================================

profit["ROA"] = (

    profit["Net_Profit"]

    /

    profit["Outstanding_Balance"]

).replace(

    [np.inf, -np.inf],

    0

).fillna(0)

# ==========================================================
# PROFIT MARGIN
# ==========================================================

profit["Profit_Margin"] = (

    profit["Net_Profit"]

    /

    (

        profit["Interest_Income"]

        +

        profit["Fee_Income"]

        +

        profit["Commission_Income"]

    )

).replace(

    [np.inf, -np.inf],

    0

).fillna(0)

print("Part 2 Completed Successfully.")
# ==========================================================
# FACT PROFITABILITY TABLE
# ==========================================================

fact_profitability = pd.DataFrame({

    "Profit_ID":
        range(
            1,
            len(profit) + 1
        ),

    "Date_Key":
        profit["Date_Key"],

    "Branch_ID":
        profit["Branch_ID"],

    "Product_ID":
        profit["Product_ID"],

    "Outstanding_Balance":
        profit["Outstanding_Balance"].round(2),

    "Interest_Income":
        profit["Interest_Income"].round(2),

    "Fee_Income":
        profit["Fee_Income"].round(2),

    "Commission_Income":
        profit["Commission_Income"].round(2),

    "Operating_Expense":
        profit["Operating_Expense"].round(2),

    "Provision_Expense":
        profit["Provision_Expense"].round(2),

    "Gross_Profit":
        profit["Gross_Profit"].round(2),

    "Tax":
        profit["Tax"].round(2),

    "Net_Profit":
        profit["Net_Profit"].round(2),

    "ROA":
        profit["ROA"].round(4),

    "Profit_Margin":
        profit["Profit_Margin"].round(4)

})

# ==========================================================
# SORT TABLE
# ==========================================================

fact_profitability = fact_profitability.sort_values(

    by="Profit_ID"

).reset_index(

    drop=True

)

print("Part 3 Completed Successfully.")
# ==========================================================
# DATA QUALITY CHECK
# ==========================================================

print("\n===================================")
print("FACT PROFITABILITY SUMMARY")
print("===================================")

print(f"Rows    : {len(fact_profitability):,}")
print(f"Columns : {len(fact_profitability.columns)}")

print("\nSummary Statistics")

print(

    fact_profitability[

        [

            "Outstanding_Balance",
            "Interest_Income",
            "Fee_Income",
            "Commission_Income",
            "Operating_Expense",
            "Provision_Expense",
            "Gross_Profit",
            "Tax",
            "Net_Profit",
            "ROA",
            "Profit_Margin"

        ]

    ].describe()

)

print("\nMissing Values")

print(

    fact_profitability

    .isnull()

    .sum()

)

print("\nFirst Five Rows")

print(

    fact_profitability.head()

)

# ==========================================================
# EXPORT
# ==========================================================

fact_profitability.to_csv(

    "fact_profitability.csv",

    index=False

)

print("\n===================================")
print("fact_profitability.csv Export Successful")
print("===================================")

print(f"Final Shape : {fact_profitability.shape}")
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# ==========================================================
# LOAD FACT TABLES
# ==========================================================

fact_profitability = pd.read_csv(
    "fact_profitability.csv"
)

fact_transaction = pd.read_csv(
    "fact_transaction.csv"
)

fact_deposit = pd.read_csv(
    "fact_deposit.csv"
)

fact_loan = pd.read_csv(
    "fact_loan.csv"
)

dim_customer = pd.read_csv(
    "dim_customer.csv"
)

dim_product = pd.read_csv(
    "dim_product.csv"
)

# ==========================================================
# TOTAL CUSTOMERS
# ==========================================================

total_customers = len(
    dim_customer
)

# ==========================================================
# PRODUCT REVENUE
# ==========================================================

product_profit = (

    fact_profitability

    .groupby(

        [
            "Date_Key",
            "Product_ID"
        ]

    )

    .agg({

        "Interest_Income":"sum",

        "Fee_Income":"sum",

        "Commission_Income":"sum",

        "Gross_Profit":"sum",

        "Net_Profit":"sum",

        "Outstanding_Balance":"sum"

    })

    .reset_index()

)

product_profit["Revenue"] = (

    product_profit["Interest_Income"]

    +

    product_profit["Fee_Income"]

    +

    product_profit["Commission_Income"]

)

# ==========================================================
# PRODUCT TRANSACTION SUMMARY
# ==========================================================

transaction_summary = (

    fact_transaction

    .groupby(

        [
            "Date_Key",
            "Product_ID"
        ]

    )

    .agg({

        "Transaction_Amount":"sum",

        "Transaction_ID":"count"

    })

    .reset_index()

)

transaction_summary.rename(

    columns={

        "Transaction_Amount":"Transaction_Value",

        "Transaction_ID":"Transaction_Count"

    },

    inplace=True

)

print("Part 1 Completed Successfully.")
# ==========================================================
# ACTIVE CUSTOMER SUMMARY
# ==========================================================

deposit_customers = (

    fact_deposit[
        [
            "Date_Key",
            "Product_ID",
            "Customer_ID"
        ]
    ]

)

loan_customers = (

    fact_loan[
        [
            "Date_Key",
            "Product_ID",
            "Customer_ID"
        ]
    ]

)

customer_usage = pd.concat(

    [

        deposit_customers,

        loan_customers

    ],

    ignore_index=True

)

active_customers = (

    customer_usage

    .groupby(

        [

            "Date_Key",

            "Product_ID"

        ]

    )

    .agg({

        "Customer_ID":"nunique"

    })

    .reset_index()

)

active_customers.rename(

    columns={

        "Customer_ID":"Active_Customers"

    },

    inplace=True

)

# ==========================================================
# MERGE ALL SOURCES
# ==========================================================

perf = (

    product_profit

    .merge(

        transaction_summary,

        on=[

            "Date_Key",

            "Product_ID"

        ],

        how="left"

    )

    .merge(

        active_customers,

        on=[

            "Date_Key",

            "Product_ID"

        ],

        how="left"

    )

)

# ==========================================================
# HANDLE MISSING VALUES
# ==========================================================

perf["Transaction_Value"] = (

    perf["Transaction_Value"]

    .fillna(0)

)

perf["Transaction_Count"] = (

    perf["Transaction_Count"]

    .fillna(0)

    .astype(int)

)

perf["Active_Customers"] = (

    perf["Active_Customers"]

    .fillna(0)

    .astype(int)

)

# ==========================================================
# PRODUCT KPI GENERATION
# ==========================================================

perf["New_Customers"] = (

    perf["Active_Customers"]

    *

    np.random.uniform(

        0.05,

        0.15,

        len(perf)

    )

).astype(int)

perf["Attrition"] = (

    perf["Active_Customers"]

    *

    np.random.uniform(

        0.01,

        0.08,

        len(perf)

    )

).astype(int)

perf["Cross_Sell_Count"] = (

    perf["Active_Customers"]

    *

    np.random.uniform(

        0.10,

        0.40,

        len(perf)

    )

).astype(int)

perf["Penetration_Rate"] = (

    perf["Active_Customers"]

    /

    total_customers

)

print("Part 2 Completed Successfully.")
# ==========================================================
# PRODUCT PERFORMANCE METRICS
# ==========================================================

perf["Average_Revenue_Per_Customer"] = (

    perf["Revenue"]

    /

    perf["Active_Customers"]

).replace(

    [np.inf, -np.inf],

    0

).fillna(0)

perf["Average_Transaction_Value"] = (

    perf["Transaction_Value"]

    /

    perf["Transaction_Count"]

).replace(

    [np.inf, -np.inf],

    0

).fillna(0)

perf["Profit_Margin"] = (

    perf["Net_Profit"]

    /

    perf["Revenue"]

).replace(

    [np.inf, -np.inf],

    0

).fillna(0)

# ==========================================================
# FACT PRODUCT PERFORMANCE TABLE
# ==========================================================

fact_product_performance = pd.DataFrame({

    "Product_Perf_ID":
        range(
            1,
            len(perf) + 1
        ),

    "Date_Key":
        perf["Date_Key"],

    "Product_ID":
        perf["Product_ID"],

    "Outstanding_Balance":
        perf["Outstanding_Balance"].round(2),

    "Revenue":
        perf["Revenue"].round(2),

    "Gross_Profit":
        perf["Gross_Profit"].round(2),

    "Net_Profit":
        perf["Net_Profit"].round(2),

    "Transaction_Value":
        perf["Transaction_Value"].round(2),

    "Transaction_Count":
        perf["Transaction_Count"],

    "Active_Customers":
        perf["Active_Customers"],

    "New_Customers":
        perf["New_Customers"],

    "Attrition":
        perf["Attrition"],

    "Cross_Sell_Count":
        perf["Cross_Sell_Count"],

    "Penetration_Rate":
        perf["Penetration_Rate"].round(4),

    "Average_Revenue_Per_Customer":
        perf["Average_Revenue_Per_Customer"].round(2),

    "Average_Transaction_Value":
        perf["Average_Transaction_Value"].round(2),

    "Profit_Margin":
        perf["Profit_Margin"].round(4)

})

# ==========================================================
# SORT TABLE
# ==========================================================

fact_product_performance = fact_product_performance.sort_values(

    by="Product_Perf_ID"

).reset_index(

    drop=True

)

print("Part 3 Completed Successfully.")
# ==========================================================
# DATA QUALITY CHECK
# ==========================================================

print("\n===================================")
print("FACT PRODUCT PERFORMANCE SUMMARY")
print("===================================")

print(f"Rows    : {len(fact_product_performance):,}")
print(f"Columns : {len(fact_product_performance.columns)}")

print("\nSummary Statistics")

print(

    fact_product_performance[

        [

            "Outstanding_Balance",
            "Revenue",
            "Gross_Profit",
            "Net_Profit",
            "Transaction_Value",
            "Transaction_Count",
            "Active_Customers",
            "New_Customers",
            "Attrition",
            "Cross_Sell_Count",
            "Penetration_Rate",
            "Average_Revenue_Per_Customer",
            "Average_Transaction_Value",
            "Profit_Margin"

        ]

    ].describe()

)

print("\nMissing Values")

print(

    fact_product_performance

    .isnull()

    .sum()

)

print("\nTop 10 Products by Revenue")

print(

    fact_product_performance

    .sort_values(

        "Revenue",

        ascending=False

    )

    .head(10)

)

print("\nFirst Five Rows")

print(

    fact_product_performance.head()

)

# ==========================================================
# EXPORT
# ==========================================================

fact_product_performance.to_csv(

    "fact_product_performance.csv",

    index=False

)

print("\n===================================")
print("fact_product_performance.csv Export Successful")
print("===================================")

print(f"Final Shape : {fact_product_performance.shape}")
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# ==========================================================
# LOAD FACT TABLES
# ==========================================================

fact_deposit = pd.read_csv(
    "fact_deposit.csv"
)

fact_loan = pd.read_csv(
    "fact_loan.csv"
)

fact_transaction = pd.read_csv(
    "fact_transaction.csv"
)

fact_npl = pd.read_csv(
    "fact_npl.csv"
)

fact_profitability = pd.read_csv(
    "fact_profitability.csv"
)

# ==========================================================
# DEPOSIT AGGREGATION
# ==========================================================

deposit_agg = (

    fact_deposit

    .groupby(

        [
            "Date_Key",
            "Branch_ID"
        ]

    )

    .agg({

        "Deposit_Balance":"sum",

        "Deposit_Amount":"sum",

        "Withdrawal_Amount":"sum",

        "Customer_ID":"nunique"

    })

    .reset_index()

)

deposit_agg.rename(

    columns={

        "Deposit_Balance":"Deposits",

        "Deposit_Amount":"Deposit_Flow",

        "Withdrawal_Amount":"Withdrawal_Flow",

        "Customer_ID":"Customers"

    },

    inplace=True

)

# ==========================================================
# LOAN AGGREGATION
# ==========================================================

loan_agg = (

    fact_loan

    .groupby(

        [
            "Date_Key",
            "Branch_ID"
        ]

    )

    .agg({

        "Outstanding_Balance":"sum",

        "Loan_Amount":"sum",

        "Interest_Income":"sum"

    })

    .reset_index()

)

loan_agg.rename(

    columns={

        "Outstanding_Balance":"Loans",

        "Loan_Amount":"Loan_Disbursement",

        "Interest_Income":"Loan_Interest"

    },

    inplace=True

)

# ==========================================================
# TRANSACTION AGGREGATION
# ==========================================================

txn_agg = (

    fact_transaction

    .groupby(

        [
            "Date_Key",
            "Branch_ID"
        ]

    )

    .agg({

        "Transaction_Amount":"sum",

        "Transaction_ID":"count"

    })

    .reset_index()

)

txn_agg.rename(

    columns={

        "Transaction_Amount":"Transaction_Value",

        "Transaction_ID":"Transaction_Count"

    },

    inplace=True

)

print("Part 1 Completed Successfully.")
# ==========================================================
# NPL AGGREGATION
# ==========================================================

npl_agg = (

    fact_npl

    .groupby(

        [
            "Date_Key",
            "Branch_ID"
        ]

    )

    .agg({

        "Outstanding":"sum",

        "Provision":"sum"

    })

    .reset_index()

)

npl_agg.rename(

    columns={

        "Outstanding":"NPL_Amount",

        "Provision":"Provision"

    },

    inplace=True

)

# ==========================================================
# PROFITABILITY AGGREGATION
# ==========================================================

profit_agg = (

    fact_profitability

    .groupby(

        [
            "Date_Key",
            "Branch_ID"
        ]

    )

    .agg({

        "Interest_Income":"sum",

        "Fee_Income":"sum",

        "Commission_Income":"sum",

        "Gross_Profit":"sum",

        "Operating_Expense":"sum",

        "Provision_Expense":"sum",

        "Tax":"sum",

        "Net_Profit":"sum"

    })

    .reset_index()

)

profit_agg["Revenue"] = (

    profit_agg["Interest_Income"]

    +

    profit_agg["Fee_Income"]

    +

    profit_agg["Commission_Income"]

)

profit_agg.rename(

    columns={

        "Net_Profit":"Profit"

    },

    inplace=True

)

# ==========================================================
# MERGE ALL SOURCES
# ==========================================================

branch_perf = (

    deposit_agg

    .merge(

        loan_agg,

        on=[

            "Date_Key",

            "Branch_ID"

        ],

        how="outer"

    )

    .merge(

        txn_agg,

        on=[

            "Date_Key",

            "Branch_ID"

        ],

        how="left"

    )

    .merge(

        npl_agg,

        on=[

            "Date_Key",

            "Branch_ID"

        ],

        how="left"

    )

    .merge(

        profit_agg,

        on=[

            "Date_Key",

            "Branch_ID"

        ],

        how="left"

    )

)

branch_perf = branch_perf.fillna(0)

print("Part 2 Completed Successfully.")
# ==========================================================
# BRANCH KPI CALCULATIONS
# ==========================================================

branch_perf["NPL_Ratio"] = np.where(

    branch_perf["Loans"] > 0,

    branch_perf["NPL_Amount"]

    /

    branch_perf["Loans"],

    0

)

branch_perf["CASA_Ratio"] = np.random.uniform(

    0.40,

    0.85,

    len(branch_perf)

)

branch_perf["Cost_To_Income"] = np.where(

    branch_perf["Revenue"] > 0,

    branch_perf["Operating_Expense"]

    /

    branch_perf["Revenue"],

    0

)

branch_perf["Profit_Margin"] = np.where(

    branch_perf["Revenue"] > 0,

    branch_perf["Profit"]

    /

    branch_perf["Revenue"],

    0

)

branch_perf["Loan_Deposit_Ratio"] = np.where(

    branch_perf["Deposits"] > 0,

    branch_perf["Loans"]

    /

    branch_perf["Deposits"],

    0

)

branch_perf["Average_Revenue_Per_Customer"] = np.where(

    branch_perf["Customers"] > 0,

    branch_perf["Revenue"]

    /

    branch_perf["Customers"],

    0

)

# ==========================================================
# FACT BRANCH PERFORMANCE TABLE
# ==========================================================

fact_branch_performance = pd.DataFrame({

    "Branch_Perf_ID":
        range(
            1,
            len(branch_perf) + 1
        ),

    "Date_Key":
        branch_perf["Date_Key"],

    "Branch_ID":
        branch_perf["Branch_ID"],

    "Deposits":
        branch_perf["Deposits"].round(2),

    "Deposit_Flow":
        branch_perf["Deposit_Flow"].round(2),

    "Withdrawal_Flow":
        branch_perf["Withdrawal_Flow"].round(2),

    "Loans":
        branch_perf["Loans"].round(2),

    "Loan_Disbursement":
        branch_perf["Loan_Disbursement"].round(2),

    "Customers":
        branch_perf["Customers"],

    "Transaction_Value":
        branch_perf["Transaction_Value"].round(2),

    "Transaction_Count":
        branch_perf["Transaction_Count"],

    "Revenue":
        branch_perf["Revenue"].round(2),

    "Gross_Profit":
        branch_perf["Gross_Profit"].round(2),

    "Profit":
        branch_perf["Profit"].round(2),

    "Provision":
        branch_perf["Provision"].round(2),

    "NPL_Amount":
        branch_perf["NPL_Amount"].round(2),

    "NPL_Ratio":
        branch_perf["NPL_Ratio"].round(4),

    "CASA_Ratio":
        branch_perf["CASA_Ratio"].round(4),

    "Cost_To_Income":
        branch_perf["Cost_To_Income"].round(4),

    "Profit_Margin":
        branch_perf["Profit_Margin"].round(4),

    "Loan_Deposit_Ratio":
        branch_perf["Loan_Deposit_Ratio"].round(4),

    "Average_Revenue_Per_Customer":
        branch_perf["Average_Revenue_Per_Customer"].round(2)

})

# ==========================================================
# SORT TABLE
# ==========================================================

fact_branch_performance = fact_branch_performance.sort_values(

    by="Branch_Perf_ID"

).reset_index(

    drop=True

)

print("Part 3 Completed Successfully.")
# ==========================================================
# DATA QUALITY CHECK
# ==========================================================

print("\n===================================")
print("FACT BRANCH PERFORMANCE SUMMARY")
print("===================================")

print(f"Rows    : {len(fact_branch_performance):,}")
print(f"Columns : {len(fact_branch_performance.columns)}")

print("\nSummary Statistics")

print(

    fact_branch_performance[

        [

            "Deposits",
            "Deposit_Flow",
            "Withdrawal_Flow",
            "Loans",
            "Loan_Disbursement",
            "Customers",
            "Transaction_Value",
            "Transaction_Count",
            "Revenue",
            "Gross_Profit",
            "Profit",
            "Provision",
            "NPL_Amount",
            "NPL_Ratio",
            "CASA_Ratio",
            "Cost_To_Income",
            "Profit_Margin",
            "Loan_Deposit_Ratio",
            "Average_Revenue_Per_Customer"

        ]

    ].describe()

)

print("\nMissing Values")

print(

    fact_branch_performance

    .isnull()

    .sum()

)

print("\nTop 10 Branches by Profit")

print(

    fact_branch_performance

    .sort_values(

        "Profit",

        ascending=False

    )

    .head(10)

)

print("\nTop 10 Branches by Deposits")

print(

    fact_branch_performance

    .sort_values(

        "Deposits",

        ascending=False

    )

    .head(10)

)

print("\nFirst Five Rows")

print(

    fact_branch_performance.head()

)

# ==========================================================
# EXPORT
# ==========================================================

fact_branch_performance.to_csv(

    "fact_branch_performance.csv",

    index=False

)

print("\n===================================")
print("fact_branch_performance.csv Export Successful")
print("===================================")

print(f"Final Shape : {fact_branch_performance.shape}")
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# ==========================================================
# LOAD TABLES
# ==========================================================

fact_loan = pd.read_csv(
    "fact_loan.csv"
)

fact_npl = pd.read_csv(
    "fact_npl.csv"
)

fact_customer_activity = pd.read_csv(
    "fact_customer_activity.csv"
)

dim_customer = pd.read_csv(
    "dim_customer.csv"
)

# ==========================================================
# NPL LOOKUP
# ==========================================================

npl_lookup = set(

    fact_npl["Loan_ID"]

)

# ==========================================================
# LATEST CUSTOMER ACTIVITY
# ==========================================================

latest_activity = (

    fact_customer_activity

    .sort_values(

        "Date_Key"

    )

    .groupby(

        "Customer_ID"

    )

    .tail(1)

    .set_index(

        "Customer_ID"

    )

)

# ==========================================================
# CUSTOMER LOOKUPS
# ==========================================================

segment_lookup = (

    dim_customer

    .set_index(

        "Customer_ID"

    )["Segment_ID"]

    .to_dict()

)

income_lookup = (

    dim_customer

    .set_index(

        "Customer_ID"

    )["Income_Band"]

    .to_dict()

)

occupation_lookup = (

    dim_customer

    .set_index(

        "Customer_ID"

    )["Occupation"]

    .to_dict()

)

# ==========================================================
# INITIALIZE OUTPUT LISTS
# ==========================================================

pd_list = []
lgd_list = []
ead_list = []

risk_score_list = []

churn_list = []

fraud_list = []

grade_list = []

ecl_list = []

snapshot_date = pd.Timestamp.today().date()

print("Part 1 Completed Successfully.")
# ==========================================================
# ENTERPRISE RISK GENERATION
# ==========================================================

for row in fact_loan.itertuples():

    customer = row.Customer_ID

    segment = segment_lookup.get(customer, 1)

    income = income_lookup.get(customer, "Middle")

    occupation = occupation_lookup.get(customer, "Private Service")

    is_npl = row.Loan_ID in npl_lookup

    # --------------------------------------------------
    # Customer Activity
    # --------------------------------------------------

    if customer in latest_activity.index:

        activity = latest_activity.loc[customer]

        satisfaction = activity["Satisfaction_Score"]

        txn_count = activity["Transaction_Count"]

        digital_usage = 0

        if activity["Mobile_App_Usage"] == "Yes":
            digital_usage += 50

        if activity["Internet_Banking_Usage"] == "Yes":
            digital_usage += 50

    else:

        satisfaction = 7

        txn_count = 15

        digital_usage = 50

    # --------------------------------------------------
    # PD
    # --------------------------------------------------

    if is_npl:

        pd_value = np.random.uniform(
            0.35,
            0.80
        )

    elif row.Days_Past_Due >= 60:

        pd_value = np.random.uniform(
            0.15,
            0.35
        )

    elif row.Days_Past_Due >= 30:

        pd_value = np.random.uniform(
            0.05,
            0.18
        )

    else:

        pd_value = np.random.uniform(
            0.005,
            0.08
        )

    if segment == 3:

        pd_value *= np.random.uniform(
            0.85,
            1.10
        )

    elif segment == 2:

        pd_value *= np.random.uniform(
            0.95,
            1.15
        )

    elif segment == 4:

        pd_value *= np.random.uniform(
            0.70,
            0.95
        )

    pd_value = min(pd_value, 1)

    # --------------------------------------------------
    # LGD
    # --------------------------------------------------

    ltv = row.LTV_Ratio

    if ltv < 0.60:

        lgd_value = np.random.uniform(
            0.15,
            0.30
        )

    elif ltv < 0.80:

        lgd_value = np.random.uniform(
            0.25,
            0.45
        )

    else:

        lgd_value = np.random.uniform(
            0.45,
            0.75
        )

    # --------------------------------------------------
    # EAD
    # --------------------------------------------------

    ead_value = row.Outstanding_Balance

    # --------------------------------------------------
    # RISK SCORE
    # --------------------------------------------------

    risk_score = (

        100
        - (pd_value * 60)
        - (lgd_value * 20)
        - min(row.Days_Past_Due,120)/6

    )

    if satisfaction < 5:

        risk_score -= 5

    if txn_count < 5:

        risk_score -= 3

    if digital_usage > 80:

        risk_score += 2

    risk_score = max(

        0,

        min(

            100,

            risk_score

        )

    )

    # --------------------------------------------------
    # CHURN
    # --------------------------------------------------

    churn = 0.05

    if satisfaction < 5:

        churn += 0.25

    if txn_count < 5:

        churn += 0.15

    if row.Days_Past_Due > 30:

        churn += 0.10

    churn = min(

        churn,

        0.95

    )

    # --------------------------------------------------
    # FRAUD
    # --------------------------------------------------

    fraud = np.random.uniform(

        0.01,

        0.08

    )

    if row.Loan_Amount > 20000000:

        fraud += 0.08

    if segment == 3:

        fraud += 0.03

    fraud = min(

        fraud,

        1

    )

    # --------------------------------------------------
    # GRADE
    # --------------------------------------------------

    if risk_score >= 85:

        grade = "A"

    elif risk_score >= 70:

        grade = "B"

    elif risk_score >= 55:

        grade = "C"

    elif risk_score >= 40:

        grade = "D"

    else:

        grade = "E"

    # --------------------------------------------------
    # ECL
    # --------------------------------------------------

    ecl = (

        pd_value

        * lgd_value

        * ead_value

    )

    # --------------------------------------------------
    # STORE VALUES
    # --------------------------------------------------

    pd_list.append(pd_value)

    lgd_list.append(lgd_value)

    ead_list.append(ead_value)

    risk_score_list.append(risk_score)

    churn_list.append(churn)

    fraud_list.append(fraud)

    grade_list.append(grade)

    ecl_list.append(ecl)
# ==========================================================
# FACT RISK SCORE TABLE
# ==========================================================

fact_risk_score = pd.DataFrame({

    "Risk_ID":
        range(
            1,
            len(fact_loan)+1
        ),

    "Date_Key":
        fact_loan["Date_Key"],

    "Customer_ID":
        fact_loan["Customer_ID"],

    "Branch_ID":
        fact_loan["Branch_ID"],

    "Product_ID":
        fact_loan["Product_ID"],

    "Loan_ID":
        fact_loan["Loan_ID"],

    "Loan_Account_No":
        fact_loan["Loan_Account_No"],

    "PD":
        np.round(
            pd_list,
            4
        ),

    "LGD":
        np.round(
            lgd_list,
            4
        ),

    "EAD":
        np.round(
            ead_list,
            2
        ),

    "Risk_Score":
        np.round(
            risk_score_list,
            2
        ),

    "Risk_Grade":
        grade_list,

    "Expected_Credit_Loss":
        np.round(
            ecl_list,
            2
        ),

    "Churn_Probability":
        np.round(
            churn_list,
            4
        ),

    "Fraud_Risk":
        np.round(
            fraud_list,
            4
        ),

    "Model_Version":
        "Risk_Model_v2.0",

    "Snapshot_Date":
        snapshot_date

})

# ==========================================================
# SORT TABLE
# ==========================================================

fact_risk_score = fact_risk_score.sort_values(

    by=[
        "Risk_ID"
    ]

).reset_index(

    drop=True

)

# ==========================================================
# DATA QUALITY CHECK
# ==========================================================

print("\n==============================")
print("FACT RISK SCORE SUMMARY")
print("==============================")

print(f"Rows    : {len(fact_risk_score):,}")
print(f"Columns : {len(fact_risk_score.columns)}")

print("\nRisk Grade Distribution")

print(

    fact_risk_score["Risk_Grade"]

    .value_counts()

)

print("\nAverage Risk Score")

print(

    round(

        fact_risk_score["Risk_Score"].mean(),

        2

    )

)

print("\nAverage PD")

print(

    round(

        fact_risk_score["PD"].mean(),

        4

    )

)

print("\nMissing Values")

print(

    fact_risk_score

    .isnull()

    .sum()

)

print("\nFirst Five Rows")

print(

    fact_risk_score.head()

)

# ==========================================================
# EXPORT
# ==========================================================

fact_risk_score.to_csv(

    "fact_risk_score.csv",

    index=False

)

print("\n===================================")
print("fact_risk_score.csv Export Successful")
print("===================================")
```
